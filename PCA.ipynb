## data Processing

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.head()
```

```python
df=data.drop(columns='Class',axis=1)
```

```python
df.head(3)
```

```python
## getting the optimal number of pca
from sklearn.decomposition import PCA
pca=PCA()
principleComponents=pca.fit_transform(df)
```

```python
principleComponents
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))

plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

```python
#From the diagram above, it can be seen that 3 principal components explain almost 95% of the variance in data

#So, instead of giving all the columns as input, weâ€™d only feed these 3 principal components of the data to the machine learning algorithm and weâ€™d obtain a similar result.
```

```python
pca=PCA(n_components=3)
new_data=pca.fit_transform(df)
new_data
```

```python
principal_df=pd.DataFrame(data=new_data
                         ,columns=['Pc1','pc2','pc3'])
```

```python
principal_df.head()
x=principal_df
y=data['Class']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
x.shape
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
r2_score(y_test,y_pred)
```
